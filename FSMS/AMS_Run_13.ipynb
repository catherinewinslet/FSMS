```python
# REGISTER USER FUNCTION ---> WHEN 'Register' BUTTON IN MainWindow IS PRESSED
# DETECTS FACE IN THE FRAME AND STORES IMAGES
def RegisterUser():
    
    l1 = RegisterIDEntry.get()
    l2 = RegisterNameEntry.get()
    if l1 and l2 == '':
        messagebox.showerror( message='Please enter ID!',icon='error')  
    elif l2 == '':
        messagebox.showerror( message='Please enter name!',icon='error')  
    else:
        try:
            # VIDEO CAPTURE AND STORE IN 'TrainingImage' FOLDER
            cam = cv2.VideoCapture(0)
            detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
            Enrollment = RegisterIDEntry.get()
            Name = RegisterNameEntry.get()
            sampleNum = 0
            while (True):
                ret, img = cam.read()
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = detector.detectMultiScale(gray, 1.3, 5)
                for (x, y, w, h) in faces:
                    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
                    # incrementing sample number
                    sampleNum = sampleNum + 1
                    # saving the captured face in the dataset folder
                    cv2.imwrite("TrainingImage/ " + Name + "." + Enrollment + '.' + str(sampleNum) + ".jpg",
                                gray[y:y + h, x:x + w])
                    cv2.imshow('Frame', img)
                # wait for 100 miliseconds
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break
                # break if the sample number is morethan 100
                elif sampleNum > 70:
                    break
            cam.release()
            cv2.destroyAllWindows()
            
            # APPEND DETAILS TO 'EmpDetails.csv' FILE
            ts = time.time()
            Date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
            row = [Enrollment, Name, Date]
            with open('EmpDetails\EmpDetails.csv', 'a+') as csvFile:
                writer = csv.writer(csvFile, delimiter=',')
                writer.writerow(row)
                csvFile.close()
            
            # CLEAR ALL FIELDS
            RegisterIDEntry.delete(first=0,last=22)
            RegisterNameEntry.delete(first=0,last=22)
            
            # STATUS MESSAGE
            displaystring.set("Images saved. Click on 'Train Images' to proceed...")
            
        except FileExistsError as F:
            messagebox.showerror( message='Employee data already exists!',icon='error') 
```

```python
# AUTOMATIC LOGIN FUNCTION ---> WHEN 'LOGIN' BUTTON IN MainWindow IS PRESSED
# PROMPTS USER TO ENTER A ROOM NUMBER
def AutomaticLogin():
    
    # AUTOMATIC FILL FUNCTION ---> WHEN 'LOGIN' BUTTON IN AutomaticLogWindow IS PRESSED
    # 1. USES TRAINED MODEL TO DETECT AND RECOGNIZE FACE
    # 2. ADD DETAILS TO LOG
    def AutomaticFill():
        RoomNum = RoomNumberEntry.get()
        now = time.time()  ###For calculate seconds of video
        future = now + 20
        if time.time() < future:
            if RoomNum == '':
                messagebox.showerror( message='Invalid room number!',icon='error') 
            else:
                recognizer = cv2.face.LBPHFaceRecognizer_create()  # cv2.createLBPHFaceRecognizer()
                try:
                    recognizer.read("TrainingImageLabel\Trainner.yml")
                except:
                    messagebox.showerror( message='Model not found, please train model first!',icon='error')

                # STATUS MESSAGE
                status3 = Label(AutomaticLogWindow, text="Please wait while scanning", font=('times',12), bg="#DEDDE5", fg='#444444')
                status3.grid(row=8, columnspan=8, sticky='W', pady=(10,0))

                # DETECTS FACIAL FEATURES
                harcascadePath = "haarcascade_frontalface_default.xml"
                faceCascade = cv2.CascadeClassifier(harcascadePath)
                df = pd.read_csv("EmpDetails\EmpDetails.csv")
                cam = cv2.VideoCapture(0)
                font = cv2.FONT_HERSHEY_SIMPLEX
                rowww = list()
                global flag
                flag = 0
                
                # READ AND PREDICT FACES IN THE FRAME
                while True:
                    ret, im = cam.read()
                    # CONVERT READ IMAGES TO GRAY SCALE
                    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
                    faces = faceCascade.detectMultiScale(gray, 1.2, 5)
                    for (x, y, w, h) in faces:
                        global Id
                        Id, conf = recognizer.predict(gray[y:y + h, x:x + w])
                        if (conf <70):
                            global room
                            global aa
                            global date
                            global timeStamp
                            room = RoomNumberEntry.get()
                            ts = time.time()
                            date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
                            timeStamp = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')
                            aa = df.loc[df['Enrollment'] == Id]['Name'].values
                            global tt
                            tt = str(Id) + "-" + aa
                            En = '15624031' + str(Id)
                            rowww = [Id, aa, date, timeStamp]
                            cv2.rectangle(im, (x, y), (x + w, y + h), (0, 260, 0), 7)
                            cv2.putText(im, str(tt), (x + h, y), font, 1, (255, 255, 0,), 4)
                            

                        else:
                            # IF MODEL CANNOT RECOGNIZE FACE
                            Id = 'Unknown'
                            flag = 1
                            tt = str(Id)
                            cv2.rectangle(im, (x, y), (x + w, y + h), (0, 25, 255), 7)
                            cv2.putText(im, str(tt), (x + h, y), font, 1, (0, 25, 255), 4)
                            
                    if time.time() > future:
                        break

                    cv2.imshow('Scanning...', im)
                    key = cv2.waitKey(10) & 0xff
                    if key == 27:
                        break

                # IF ANY UNKNOWN PERSON IS PRESENT, DENY ACCESS ELSE GRANT ACCESS
                if flag == 1:
                    cam.release()
                    cv2.destroyAllWindows()
                    AutomaticLogWindow.destroy()
                    messagebox.showerror( message='Access Denied! Please register before entering.',icon='error')
                    return
                else:
                    cam.release()
                    cv2.destroyAllWindows()
                    messagebox.showinfo( message='Access Granted',icon='info')
                    
                    # WRITE FILE NAME USING CURRENT YEAR AND MONTH
                    x = datetime.datetime.now()
                    AutomaticLogFileName = 'LogBook/' + str(x.year) + "_" + str(x.month) + '.csv'
                    
                    # CHECK IF FILE WITH THE NAME 'yyyy_mm.csv' EXISTS
                    if os.path.isfile(AutomaticLogFileName):
                        with open(AutomaticLogFileName,'a+') as csv_file:
                            csv_file.close()
                    else:
                        with open(AutomaticLogFileName,'w') as csv_file:
                            csv_file.close()
                    
                    # WRITE DETAILS INTO CSV FILE
                    try:
                        with open(AutomaticLogFileName, "a+",newline='') as csv_file:
                            csv_writer = csv.writer(csv_file)
                            writer = csv.writer(csv_file, delimiter=',')
                            writer.writerow(rowww)
                            csv_file.close()
                    except Exception as e:
                        messagebox.showinfo(message=e, icon ='info')
                        print(e)
                        return

                    ####### create status bar  for This#####

                AutomaticLogWindow.destroy()

                # DISPLAY LOG RECORDED
                root = tk.Tk()
                root.title("Log for " + room)
                with open(AutomaticLogFileName, newline="") as file:
                    reader = csv.reader(file)
                    r = 0
                    for col in reader:
                        c = 0
                        for row in col:
                            label = tk.Label(root, width=11, font=('times', 15), text=row)
                            label.grid(row=r, column=(c*2),padx=5,pady=5)
                            if(c*2+1 != 5):
                                ttk.Separator(root, orient=VERTICAL).grid(column=(c*2)+1, row=r, sticky='ns')
                            c += 1
                        r += 1
                root.mainloop()

    # AUTOMAITC LOGIN WINDOW
    AutomaticLogWindow = tk.Tk()
    AutomaticLogWindow.iconbitmap('AMS.ico')
    AutomaticLogWindow.title("FSMS - Login")
    AutomaticLogWindow.configure(bg='#DEDDE5')
    AutomaticLogWindow.geometry("+{}+{}".format(positionRight-300, positionDown-100))
    
    # LABEL WIDGET ---> "Room Number"
    RoomNumberLabel = tk.Label(AutomaticLogWindow, text="Room Number", bg="#DEDDE5", font=('times', 15))
    RoomNumberLabel.grid(row=0, column=0, sticky=('W','S'), padx=30, pady=(20,0))

    # TEXT ENTRY WIDGET ---> TAKES INTEGER INPUT
    RoomNumberEntry = tk.Entry(AutomaticLogWindow, validate="key", width=20, font=('times', 23))
    RoomNumberEntry['validatecommand'] = (RoomNumberEntry.register(testVal),'%P','%d')
    RoomNumberEntry.grid(row=1, column=0, padx=30, pady=(0,20), sticky=('wn'))

    # AUTOMATIC LOGIN BUTTON WIDGET ---> CALLS 'AutomaticFill' FUNCTION
    AutomaticLoginButton = tk.Button(AutomaticLogWindow, text="Login", command=AutomaticFill, width=13, activebackground='#999999', bg='#8894C3', borderwidth=1, font=('times', 15))
    AutomaticLoginButton.grid(row=2, column=0, sticky='E', padx=30, pady=(0,20))

    AutomaticLogWindow.mainloop()


########################################################################################################################### 
    

# EMPLOYEE RECORDS FUNCTION ---> WHEN 'EMP RECORDS' BUTTON IN MainWindow IS PRESSED
# PROMPTS USER TO ENTER ADMIN USERNAME & PASSWORD
def EmployeeRecords():
    win = tk.Tk()
    win.iconbitmap('AMS.ico')
    win.title("FSMS - Admin Login")

    # VALIDATION FUNCTION ---> WHEN 'EMP RECORDS' BUTTON IN MainWindow IS PRESSED
    # CHECKS ENTERED ADMIN USERNAME & PASSWORD AND DISPLAY EMPLOYEE RECORDS
    def Validate():
        username = UsernameEntry.get()
        password = PasswordEntry.get()

        if username == 'fsms' :
            if password == 'fsms':
                win.destroy()

                # DISPLAY EMP-RECORDS WINDOW
                root = tk.Tk()
                root.title("FSMS - Employee Details")
                root.configure(height=10)

                # DISPLAY COLUMN HEADERS
                emp = tk.Label(root,width=11,font=('times',15,'bold'),text = "ID")
                ttk.Separator(root, orient=VERTICAL).grid(column=1, row=0, sticky='ns')
                nm = tk.Label(root,width=11,font=('times',15,'bold'),text = "NAME")
                ttk.Separator(root, orient=VERTICAL).grid(column=3, row=0, sticky='ns')
                dt = tk.Label(root,width=11,font=('times',15,'bold'),text = "DATE")
                emp.grid(column=0,row=0,padx=5,pady=5)
                nm.grid(column=2,row=0,padx=5,pady=5)
                dt.grid(column=4,row=0,padx=5,pady=5)
                
                #DISPLAY CONTENTS OF CSV FILE IN TABLE FORMAT
                cs = 'E:/FSMS/EmpDetails/EmpDetails.csv'
                with open(cs, newline="") as file:
                    reader = csv.reader(file)
                    r = 1
                    for col in reader:
                        c = 0
                        for row in col:
                            label = tk.Label(root, width=11, font=('times', 15), text=row)
                            label.grid(row=r, column=(c*2),padx=5,pady=5)
                            if(c*2+1 != 5):
                                ttk.Separator(root, orient=VERTICAL).grid(column=(c*2)+1, row=r, sticky='ns')
                            c += 1
                        r += 1
                        
                root.grid_rowconfigure(0, weight=1)
                root.grid_columnconfigure(0, weight=1)
                root.mainloop()
            else:
                messagebox.showerror( message='Incorrect password!',icon='error') 
        else:
            messagebox.showerror( message='Incorrect username!',icon='error')

    # LABEL WIDGET ---> "Enter Username"
    UsernameLabel = tk.Label(win, text="Enter username", font=('times', 15))
    UsernameLabel.grid(row=0, column=0, sticky=('S','W'), padx=40, pady=(10,0))

    # LABEL WIDGET ---> "Enter Password"
    PasswordLabel = tk.Label(win, text="Enter password", font=('times', 15))
    PasswordLabel.grid(row=2, column=0, sticky=('S','W'), padx=40, pady=(10,0))

    # CLEAR CONTENTS OF TEXT FIELDS
    def c00():
        UsernameEntry.delete(first=0, last=22)
    def c11():
        PasswordEntry.delete(first=0, last=22)
        
    # TEXT ENTRY WIDGET ---> TAKES STRING INPUT (fsms)
    UsernameEntry = tk.Entry(win, width=20, font=('times', 18))
    UsernameEntry.grid(row=1, column=0, padx=(40,0), pady=(5,5))

    # TEXT ENTRY WIDGET ---> TAKES ALPHANUM INPUT (fsms)
    PasswordEntry = tk.Entry(win, width=20,show="*",  font=('times', 18))
    PasswordEntry.grid(row=3,column=0,padx=(40,0),pady=(5,10))
    
    # CLEAR BUTTON ---> CLEARS TEXT FIELDS
    c0 = tk.Button(win, text="Clear", command=c00,activebackground='#999999',bg='#8894C3',borderwidth=0,font=('times',15))
    c1 = tk.Button(win, text="Clear", command=c11,activebackground='#999999',bg='#8894C3',borderwidth=0,font=('times',15))
    c0.grid(row=1,column=1,sticky=('W'),pady=(5,5),padx=(0,40))
    c1.grid(row=3,column=1,sticky=('W'),pady=(5,10),padx=(0,40))

    # EMP-RECORDS LOGIN BUTTON WIDGET ---> CALLS 'Validate' FUNCTION
    EmpRecordsLoginButton = tk.Button(win, text="Login", command=Validate, width=13, activebackground='#999999', bg='#8894C3', borderwidth=1, font=('times', 15))
    EmpRecordsLoginButton.grid(row=4, column=0, padx=20, pady=20)
    
    # STATUS BAR
    status = Label(win, textvariable=displaystring, font=('times',12), fg='#444444')
    status.grid(row=8, columnspan=8, sticky='W', pady=(10,0))
    
    win.mainloop()
```

```python
# VISITOR REGISTER FUNCTION ---> WHEN 'Register' IN VISITOR WINDOW 'vis'
# DETECTS AND STORES FACES IN CSV FILE
def VisitorRegister():
    
    s1 = VisitorNameEntry.get()
    s2 = ID.get()
    if s1 == '':
        messagebox.showerror( message='Please enter ID!',icon='error')  
    elif s2 == '':
        messagebox.showerror( message='Please enter VisitorNameEntry!',icon='error')
    else:
        try:
            cam = cv2.VideoCapture(0)
            detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
            Enrollment = ID.get()
            Name = VisitorNameEntry.get()
            sampleNum = 0
            while (True):
                ret, img = cam.read()
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = detector.detectMultiScale(gray, 1.3, 5)
                for (x, y, w, h) in faces:
                    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
                    # incrementing sample number
                    sampleNum = sampleNum + 1
                    # saving the captured face in the dataset folder
                    cv2.imwrite("TrainingImage/ " + Name + "." + Enrollment + '.' + str(sampleNum) + ".jpg",
                                gray[y:y + h, x:x + w])
                    cv2.imshow('Frame', img)
                # wait for 100 miliseconds
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break
                # break if the sample number is morethan 100
                elif sampleNum > 70:
                    break
            cam.release()
            cv2.destroyAllWindows()
            ts = time.time()
            Date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
            Time = datetime.datetime.fromtimestamp(ts).strftime('%H:%M:%S')
            row = [Enrollment, Name, Date, Time]
            with open('VisitorDetails\VisitorDetails.csv', 'a+') as csvFile:
                writer = csv.writer(csvFile, delimiter=',')
                writer.writerow(row)
                csvFile.close()
            #status bar "click on'train model'"
            
            
        except FileExistsError as F:
            messagebox.showerror( message='Employee data already exists!',icon='error')

# TRAIN MODEL FUNCTION ---> WHEN 'Train Model' BUTTON IN VISITOR WINDOW IS PRESSED
# 1. RETRIEVES IMAGES AND LABELS(IMAGES NAMES) FROM TRAINING IMAGES
# 2. TRAINS THE MODEL WITH THE DATA
# 3. STORES IN TRAINNER.YML
def VisitorTrainModel():
    recognizer = cv2.face.LBPHFaceRecognizer_create()
    global detector
    detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    try:
        global faces,Id
        faces, Id = getImagesAndLabels("TrainingImage")
    except Exception as e:
        messagebox.showerror( message="Please create 'TrainingImage' folder!",icon='error')

    recognizer.train(faces, np.array(Id))
    try:
        recognizer.save("TrainingImageLabel\Trainner.yml")
    except Exception as e:
        messagebox.showerror( message="Please create 'TrainingImageLabel' folder!",icon='error')

    displaystring.set('Model trained')
    status.configure(fg='green')

def getImagesAndLabels(path):
    imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
    # create empth face list
    faceSamples = []
    # create empty ID list
    Ids = []
    # now looping through all the image paths and loading the Ids and the images
    for imagePath in imagePaths:
        # loading the image and converting it to gray scale
        pilImage = Image.open(imagePath).convert('L')
        # Now we are converting the PIL image into numpy array
        imageNp = np.array(pilImage, 'uint8')
        # getting the Id from the image

        Id = int(os.path.split(imagePath)[-1].split(".")[1])
        # extract the face from the training image sample
        faces = detector.detectMultiScale(imageNp)
        # If a face is there then append that in the list as well as Id of it
        for (x, y, w, h) in faces:
            faceSamples.append(imageNp[y:y + h, x:x + w])
            Ids.append(Id)
    return faceSamples, Ids

# VISITOR LOGIN FUNCTION ---> WHEN 'Login' BUTTON IN 'vis' WINDOW IS PRESSED
# 1. USES TRAINED MODEL TO DETECT AND RECOGNIZE FACE
# 2. ADD DETAILS TO LOG
def VisitorLogin():
    print("not implemented yet")

# VISITOR FUNCTION ---> WHEN 'Visitor' BUTTON IN 'MainWindow' IS PRESSED
# PRESENTS USER WITH TWO CHOICES;
# 1. LOGIN AS REGISTERED VISITOR
# 2. REGISTER AS NEW VISITOR
def visitor():
    global VisitorNameEntry,ID
    
    # VISITOR WINDOW
    vis = tk.Tk()
    vis.title("FSMS - Visitor Registry")
    vis.configure(bg="#DEDDE5")
    
    # CLEAR FUNCTION
    def clear():
        VisitorNameEntry.delete(first=0, last=22)
    def clear1():
        ID.delete(first=0, last=22)
    def clear2():
        RoomsAccessibleEntry.delete(first=0, last=22)
    
    # CLEAR BUTTON ---> CALLS CLEAR FUNCTION
    clearButton = tk.Button(vis,text="clear",command=clear,activebackground='#999999',bg='#8894C3',borderwidth=0,font=('times',15))
    clearButton.grid(row=4,column=1,sticky='nw')
    clearButton1 = tk.Button(vis,text="clear",command=clear1,activebackground='#999999',bg='#8894C3',borderwidth=0,font=('times',15))
    clearButton1.grid(row=6,column=1,sticky='nw')
    clearButton2 = tk.Button(vis,text="clear",command=clear2,activebackground='#999999',bg='#8894C3',borderwidth=0,font=('times',15))
    clearButton2.grid(row=8,column=1,sticky='nw')
     
    # VISITOR LOGIN BUTTON WIDGET ---> CALLS 'VisitorLogin' FUNCTION
    VisitorLoginButton = tk.Button(vis, text="Login", width=13, command=VisitorLogin, activebackground='#999999', bg='#8894C3',borderwidth=1,font=('times', 15))
    VisitorLoginButton.grid(row=0, column=0, columnspan=3, pady=25)
    
    ttk.Separator(vis, orient=HORIZONTAL).grid(column=0, row=1, columnspan=2, sticky='ew',pady=10,padx=10)

    # VISITOR NAME LABEL WIDGET ---> "Name"
    VisitorNameLabel = tk.Label(vis, text="Name",bg="#DEDDE5", font=('times', 15))
    VisitorNameLabel.grid(row=3, column=0, sticky=('W','S'), padx=(20,0), pady=(15,0))

    # TEXT ENTRY WIDGET ---> TAKES STRING INPUT
    VisitorNameEntry = tk.Entry(vis, width=20, bg="white", font=('times', 20))
    VisitorNameEntry.grid(row=4, column=0, sticky=('N','W'), padx=(20,0), pady=(0,10))
    
    # ID LABEL WIDGET ---> "Identification No."
    VisitorIDEntry = tk.Label(vis, text="Identification No.",bg="#DEDDE5", font=('times', 15))  #generate ID instead
    VisitorIDEntry.grid(row=5, column=0,sticky=('W','S'),padx=(20,0),pady=(10,0))

    # TEXT ENTRY WIDGET ---> TAKES INTEGER INPUT
    ID = tk.Entry(vis, validate="key", width=20, bg="white", font=('times', 20))
    ID['validatecommand'] = (ID.register(testVal),'%P','%d')
    ID.grid(row=6, column=0, sticky=('W','N'), padx=(20,0), pady=(0,10))

    # ROOMS LABEL WIDGET 1 ---> "Rooms Accessible"
    RoomsLabel1 = tk.Label(vis, text="Rooms Accessible", bg="#DEDDE5", font=('times', 15))
    RoomsLabel1.grid(row=7, column=0, sticky=('W','S'), padx=(20,0), pady=(10,0))
    
    # ROOMS LABEL WIDGET 2 ---> "separated by commas. e.g. (123, 101, 222)"
    RoomsLabel2 = tk.Label(vis, text="separated by commas. e.g (123, 101, 222)", bg="#DEDDE5", font=('times', 10))
    RoomsLabel2.grid(row=9, column=0, sticky=('N','W'), padx=20, pady=(0,20))

    # TEXT ENTRY WIDGET ---> TAKES COMMA-SEPARATED INTERGER VALUES
    RoomsAccessibleEntry = tk.Entry(vis, width=20, bg="white", font=('times', 20))
    RoomsAccessibleEntry.grid(row=8, column=0, sticky=('W','N'), padx=(20,0))
    
    # VISITOR REGISTER BUTTON WIDGET ---> CALLS 'VisitorRegister' FUNCTION
    VisitorRegisterButton = tk.Button(vis, text="Register", width=13, command=VisitorRegister, activebackground='#999999', bg='#8894C3',borderwidth=1,font=('times', 15))
    VisitorRegisterButton.grid(row=10, column=0, pady=(10,20))

    # VISITOR TRAIN MODEL BUTTON WIDGET ---> CALLS 'VisitorTrainModel' FUNCTION
    VisitorRegisterButton = tk.Button(vis, text="Train Model", width=13, command=VisitorTrainModel, activebackground='#999999',bg='#8894C3',borderwidth=1,font=( 'times',15))
    VisitorRegisterButton.grid(row=10, column=1, pady=(10,20), padx=(10,40))
    
    # STATUS BAR
    displaystring1=StringVar()
    status1 = Label(vis, textvariable=displaystring1,font=('times',12),bg="#DEDDE5",fg='#444444')
    status1.grid(row=14,columnspan=8,sticky='W',pady=(10,0))
    displaystring1.set("Click 'Register' to Add new Visitor")

    vis.resizable(False, False)
    vis.iconbitmap('AMS.ico')
    vis.mainloop()
```
