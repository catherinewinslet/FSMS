```python
#displays log for the month
def log_book():
    import csv
    import os
    import re 
    import tkinter as tk
    from tkinter import ttk
    logwindow = tk.Tk()
    logwindow.title("FSMS - Log Book")
    logwindow.configure(height=10)
    
    roottt = Listbox(logwindow, height=5)
    roottt.grid(column=0, row=0, sticky=(N,W,E,S))
    s = ttk.Scrollbar(logwindow, orient=VERTICAL, command=roottt.yview)
    s.grid(column=1, row=0, sticky=(N,S))
    roottt['yscrollcommand'] = s.set
    
    tree = ttk.Treeview(logwindow, height=7)
    tree.heading("#0",text="Year",anchor=tk.W)
    tree.grid(row=0,column=0)
    tree.column('#0', width=400)
    
    s = ttk.Scrollbar(logwindow, orient=VERTICAL, command=tree.yview)
    s.grid(column=1, row=0, sticky=(N,S))
    tree['yscrollcommand'] = s.set
    s = ttk.Scrollbar(logwindow, orient=HORIZONTAL, command=tree.xview)
    s.grid(column=0, row=1, sticky=(W,E))
    tree['xscrollcommand'] = s.set
    
    months={1:'January',
            2:'Febuary',
            3:'March',
            4:'April',
            5:'May',
            6:'June',
            7:'July',
            8:'August',
            9:'September',
            10:'October',
            11:'November',
            12:'December'}
    
    path = 'LogBook'
    for filename in os.listdir(path):
        if (filename.endswith('.csv')):
            try:
                tree.insert('','end',filename[:4],text=filename[:4])
            except:
                print("")
            Reg = re.compile(r'.*_(.*).csv')
            mo = Reg.search(filename)
            tree.insert(filename[:4], 'end', text=months[int(mo.group(1))])
    
    def OnDoubleClick(event):
        item = tree.identify('item',event.x,event.y)
        monthnumber = ((list(months.keys())[list(months.values()).index(tree.item(item,"text"))]))
        csvfilename = tree.parent(item)+'_'+str(monthnumber)+'.csv'
        
        monthlog = tk.Tk()
        title = 'FSMS - Showing Log for ' + tree.item(item,"text")
        monthlog.title(title)
        monthlog.configure(height=10)
        
        emp = tk.Label(monthlog,width=11,font=('times',15,'bold'),text = "ID")
        ttk.Separator(monthlog, orient=VERTICAL).grid(column=1, row=0, sticky='ns')
        nm = tk.Label(monthlog,width=11,font=('times',15,'bold'),text = "NAME")
        ttk.Separator(monthlog, orient=VERTICAL).grid(column=3, row=0, sticky='ns')
        dt = tk.Label(monthlog,width=11,font=('times',15,'bold'),text = "DATE")
        ttk.Separator(monthlog, orient=VERTICAL).grid(column=5, row=0, sticky='ns')
        tm = tk.Label(monthlog,width=11,font=('times',15,'bold'),text = "TIME")
        emp.grid(column=0,row=0,padx=5,pady=5)
        nm.grid(column=2,row=0,padx=5,pady=5)
        dt.grid(column=4,row=0,padx=5,pady=5)
        tm.grid(column=6,row=0,padx=5,pady=5)
        
        cs = 'LogBook/'+csvfilename
        with open(cs, newline="") as file:
            reader = csv.reader(file)
            r = 1
            for col in reader:
                c = 0
                for row in col:
                    label = tk.Label(monthlog, width=11, font=('times', 15), text=row)
                    label.grid(row=r, column=(c*2),padx=5,pady=5)
                    if(c*2+1 != 7):
                        ttk.Separator(monthlog, orient=VERTICAL).grid(column=(c*2)+1, row=r, sticky='ns')
                    c += 1
                r += 1
        monthlog.mainloop()  
        
    tree.bind("<Double-1>", OnDoubleClick)
            
    ttk.Sizegrip(logwindow).grid(column=1, row=1, sticky=(S,E))
    logwindow.grid_rowconfigure(0, weight=1)
    logwindow.grid_columnconfigure(0, weight=1)
    logwindow.mainloop()
```
